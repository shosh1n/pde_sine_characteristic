```python
x = np.arange(0,0.001,0.1/41000)
vis = 0.180
t = 1.8
n = np.arange(1,20,1)
coeff =[0 for i in range(np.size(n))]
cutCoeff  =[0 for i in range(np.size(n))]
w=2*np.pi
f1=20000
f2=18000
N=80000
```

```python
for i in range(np.size(n)):
        coeff[i]= integrate.quad(lambda x: np.exp(-1/(4*vis*np.pi)*(1-np.cos(x)))*np.cos(i*x),\
        0,np.pi*2)
```

```python
a0 = integrate.quad(lambda x: np.exp((-1/(4*vis*np.pi*1))*(1-np.cos(x))), 0,2*np.pi)
```

```python
for i in range(np.size(n)):
    cutCoeff[i] = coeff[i][0]
```

```python
def u(x): 
    out = (2 +4*np.pi*vis\
           *((cutCoeff[0]*np.exp(-vis*1*1*t)*1*np.sin(1*x*f1*w)\
           + cutCoeff[1]*np.exp(-vis*2*2*t)*2*np.sin(2*x*f1*w)\
           + cutCoeff[2]*np.exp(-vis*3*3*t)*3*np.sin(3*x*f1*w)\
           + cutCoeff[3]*np.exp(-vis*4*4*t)*4*np.sin(4*x*f1*w)\
           + cutCoeff[4]*np.exp(-vis*5*5*t)*5*np.sin(5*x*f1*w))\
                                    /                         \
          (a0+(cutCoeff[0]*np.exp(-vis*1*1*t)*1*np.cos(1*x*f1*w)\
          +cutCoeff[1]*np.exp(-vis*2*2*t)*2*np.cos(2*x*f1*w)\
          +cutCoeff[2]*np.exp(-vis*3*3*t)*3*np.cos(3*x*f1*w)\
          +cutCoeff[3]*np.exp(-vis*4*4*t)*4*np.cos(4*x*f1*w)\
          +cutCoeff[4]*np.exp(-vis*5*5*t)*5*np.cos(5*x*f1*w))))) \
                +                       \
            (2 +4*np.pi*vis\
           *((cutCoeff[0]*np.exp(-vis*1*1*t)*1*np.sin(1*x*f2*w)\
           + cutCoeff[1]*np.exp(-vis*2*2*t)*2*np.sin(2*x*f2*w)\
           + cutCoeff[2]*np.exp(-vis*3*3*t)*3*np.sin(3*x*f2*w)\
           + cutCoeff[3]*np.exp(-vis*4*4*t)*4*np.sin(4*x*f2*w)\
           + cutCoeff[4]*np.exp(-vis*5*5*t)*5*np.sin(5*x*f2*w))\
                                    /                         \
          (a0+(cutCoeff[0]*np.exp(-vis*1*1*t)*1*np.cos(1*x*f2*w)\
           + cutCoeff[1]*np.exp(-vis*2*2*t)*2*np.cos(2*x*f2*w)\
           + cutCoeff[2]*np.exp(-vis*3*3*t)*3*np.cos(3*x*f2*w)\
           + cutCoeff[3]*np.exp(-vis*4*4*t)*4*np.cos(4*x*f2*w)\
           + cutCoeff[4]*np.exp(-vis*5*5*t)*5*np.cos(5*x*f2*w))))) 
    
    return out
```
