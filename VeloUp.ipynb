```python
import numpy as np
import itertools
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import FuncFormatter, MultipleLocator, LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import figure
from matplotlib import gridspec
#%matplotlib notebook
```

```python
x = np.arange(0,2*np.pi,4*np.pi/100)
t = np.arange(0,1.5,1.5/100) #Schritte auf y bis 1.5
tenx =  np.arange(0,2*np.pi,2*np.pi/50) #Schritte auf x bis 2*PI

xt = np.zeros((np.size(tenx),np.size(t))) #Array mit der Breite 50 und Tiefe 100
```

```python
lolx = [
        xt[i] + tenx[i] for i in range(np.size(tenx)) #Temp-Array #50 Breite 100 Tiefe(selber Wert) bis
        ]                                             #50 mal bis 2*PI wobei einer der Schritte bis 2 Pi
                                                      #100 Mal vorhanden.
```

```python
sinx =[np.sin(lolx[i]) for i in range(np.size(tenx))] #Sinus Graph mit Array erzeugt. 100mal gleicher sin wert tief
w, h = np.size(tenx), np.size(t)
gradx = [[0 for i in range(h)] for g in range(w)]  #leeres Array für die Gerade berechnung                                                    
print("hi")
```

```python
for i in range(len(sinx)):                            #Geraden-Funktion des Arrays wird erstellt.
    for j in range(np.size(t)):                       #jeder der 50 Achsenabs      chnitte bekommt durch die Zeit 
        gradx[i][j] = [sinx[i][j]*t[j] +lolx[i][j]]   #rattert die 50-werte die 100 tief sind mit zunehmender Zeit ab: Geradengleichung
```

```python
timex = [[0 for i in range(w)] for g in range(h)]
timecons = [[0 for i in range(w)] for g in range(h)]
```

```python
for k in range(np.size(t)):                #erstellt hundertmal ein fünfzigfach tiefes Array
    for q in range(np.size(tenx)):
        timex[k][q] = gradx[q][k]   #gradx hat 50 Arrays mit 100 Werten - timex 100 Arrays mit 50 Werten
                                    #Die 100 Werte des ersten Arrays(gradx) sind nun
                                    #die jeweils ersten Werte der 100 Arrays(timex)
```

```python
timeCord = [0 for g in range(np.size(tenx))]

for i in range(np.size(timeCord)):
           timeCord[i] = timex[99][i]

yoor = 2+np.sin(tenx)
```

```python
fig = plt.figure()
fig, ax = plt.subplots()
print(len(gradx))
plt.plot(timeCord, yoor)

ax.xaxis.set_major_formatter(FuncFormatter(
   lambda val,pos: '{:.0g}$\pi$'.format(val/np.pi) if val !=0 else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi))
ax.set(xlabel='Ort (x)', ylabel='Geschwindigkeit (v)',
       title='Schockwelle - Sinus')

plt.show()
```
