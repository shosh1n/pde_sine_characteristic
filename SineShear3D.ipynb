```python
import numpy as np
import itertools
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import FuncFormatter, MultipleLocator, LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
x = np.arange(0,2*np.pi,2*np.pi/100)
t = np.arange(0,1.5,1.5/100)
tenx =  np.arange(0,2*np.pi,2*np.pi/50)

xt = np.zeros((np.size(tenx),np.size(t)))
```

```python
lolx = [
        xt[i] + tenx[i] for i in range(np.size(tenx))
        ]
```

```python
sinx =[np.sin(lolx[i]) for i in range(np.size(tenx))]
```

```python
gradx = [sinx[i]*t +tenx[i]  for i in range(np.size(tenx))]
```

```python
w, h = np.size(tenx), np.size(t)
timex = 0
```

```python
timex = [[0 for i in range(w)] for g in range(h)]
timecons = [[0 for i in range(w)] for g in range(h)]
```

```python
for k in range(np.size(t)):                #erstellt hundertmal ein f√ºnfzigfach tiefes Array
    for q in range(np.size(tenx)):
        timex[k][q] = gradx[q][k]   #gradx hat 50 Arrays mit 100 Werten - timex 100 Arrays mit 50 Werten
                                    #Die 100 Werte des ersten Arrays(gradx) sind nun
                                    #die jeweils ersten Werte der 100 Arrays(timex)
                                    
  
```

```python
for k in range(np.size(t)):
    for q in range(np.size(tenx)):
        timecons[k][q] = 2 + np.sin(timex[k][q])
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

color=iter(cm.twilight(np.linspace(0,1,50)))

for i in range(np.size(tenx)):
        c=next(color)  
        ax.plot3D(gradx[i],t,timecons[0][i], c=c)
          
ax.view_init(13, 90)    
ax.xaxis.set_major_formatter(FuncFormatter(
   lambda val,pos: '{:.0g}$\pi$'.format(val/np.pi) if val !=0 else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi))
ax.set(xlabel='Ort (x)', ylabel='Zeit (t)', zlabel='Geschwindigkeit',
       title='Wellenaufsteilung- [t] gro√ü - Sinus')
fig.savefig("Wellenaufsteilung-Sinus_t_gro√ü2.png", dpi=300)
plt.show()
```
